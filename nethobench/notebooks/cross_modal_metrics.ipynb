# NethoBench Cross-Modal Full Analysis
This notebook evaluates **neural**, **behavior**, and **cross-modal** coupling metrics on multimodal CSVs.
- Configure paths and config JSON below.
- Metrics mirror the CLI: neuro/etho cores + cross-modal alignment (CCA, predictive R^2 similarity, lead/lag).
- Figures saved when run via the headless runner (plt.show is patched).

```python
from pathlib import Path
import json
import matplotlib.pyplot as plt
import pandas as pd
from nethobench.cross import compute_cross_scores

# Set these paths (overwritten by the headless runner)
gt_path = Path("/path/to/gt_multimodal.csv")
pred_path = Path("/path/to/pred_multimodal.csv")
config_path = Path("config.json")

cfg = json.loads(config_path.read_text())
scores = compute_cross_scores(pred_path, gt_path, config_path)
scores
```

```python
# Pretty-print composites
print('Neuro composite:', scores['neuro_composite'])
print('Etho composite :', scores['etho_composite'])
print('Cross composite:', scores['cross_composite'])
print('Final composite:', scores['composite'])
```

```python
# Example quick-look plot: cross-modal CCA comparison
cca = scores['cross_scores']
plt.figure(figsize=(4,3))
plt.bar(['GT CCA', 'Pred CCA'], [cca.get('cca_gt', float('nan')), cca.get('cca_pred', float('nan'))], color=['#4caf50','#2196f3'])
plt.ylabel('Mean canonical corr')
plt.title('Neural–Behavior CCA (GT vs Pred)')
plt.tight_layout()
plt.show()
```

```python
# Save scores to JSON (optional)
out_json = Path('cross_scores.json')
out_json.write_text(json.dumps(scores, indent=2))
print('Saved to', out_json.resolve())
```
